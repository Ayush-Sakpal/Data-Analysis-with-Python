# Data visualization with Seaborn

Seaborn is a python visualization library based on matplotlib that provides a high-level interface for drawing interactive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code.

```python
import seaborn as sns
```

## Basic plotting with seaborn

```python
tips = sns.load_dataset('tips')
tips
```

```python
# Create a scatter plot 
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter plot of Total Bill vs Tip')
plt.show()
```

```python
## Line plot

sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Line plot of Size vs Total Bill')
plt.show()
```

```python
## Categorical plots

# Bar plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar plot of total Bill by day')
plt.show()
```

```python
# Box plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box plot of total bill by day')
plt.show()
```

```python
# Violin Plot

sns.violinplot(x='day', y='total_bill', data=tips, color='pink')
plt.title('Violin plot')
plt.show()
```

```python
# Histogram
sns.histplot(tips['total_bill'], bins=10, kde=True)
```

```python
# KDE plot
sns.kdeplot(tips['total_bill'], fill=True)
```

```python
# Pair plot
sns.pairplot(tips)
```

```python
# Heatmap
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
